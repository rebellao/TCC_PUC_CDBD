```python
import seaborn as sns

# Mapa de correlação de Pearson, incluindo todas as cidades disponíveis no dataset
correlation_matrix_all_cities = df_p[['prod-preco', 'spo-temp', 'spo-umid', 'spo-uv', 
                                    'bra-temp', 'bra-umid', 'bra-uv',
                                    'pve-temp', 'pve-umid', 'pve-uv',
                                    'flo-temp', 'flo-umid', 'flo-uv',
                                    'rj-temp', 'rj-umid', 'rj-uv']].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix_all_cities, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de Calor das Correlações entre Preço e Fatores Climáticos (Todas as Cidades)')
plt.show()
```

```python
# calcular a correlação de Pearson por produto
produtos_correlation = []

for produto, data in df_p.groupby('prod-descr'):
    corr_values = data[['prod-preco', 'spo-temp', 'spo-umid', 'spo-uv', 
                        'bra-temp', 'bra-umid', 'bra-uv', 
                        'pve-temp', 'pve-umid', 'pve-uv', 
                        'flo-temp', 'flo-umid', 'flo-uv', 
                        'rj-temp', 'rj-umid', 'rj-uv']].corr()['prod-preco']
    

    corr_values = corr_values.drop('prod-preco')  # Remover a correlação com o próprio preço
    
    strong_correlations = corr_values[(corr_values >= 0.8) | (corr_values <= -0.8)]
    
    if not strong_correlations.empty:
        produtos_correlation.append({
            'Produto': produto,
            'Fatores Climáticos': strong_correlations.index.tolist(),
            'Correlações': strong_correlations.values.tolist()
        })

df_correlation = pd.DataFrame(produtos_correlation)
df_correlation.head()
```

```python
top_products_by_city = {}

cities = ['spo-temp', 'bra-temp', 'pve-temp', 'flo-temp', 'rj-temp']

for city in cities:
    city_correlations = []
    for index, row in df_correlation.iterrows():
        if city in row['Fatores Climáticos']:
            city_index = row['Fatores Climáticos'].index(city)
            city_correlations.append({
                'Produto': row['Produto'],
                'Cidade': city,
                'Correlacao': row['Correlações'][city_index]
            })
    
    df_city_correlations = pd.DataFrame(city_correlations)
    
    df_top_5_city_correlations = df_city_correlations.sort_values(by='Correlacao', ascending=False).head(5)
    top_products_by_city[city] = df_top_5_city_correlations

df_top_5_all_cities = pd.concat(top_products_by_city.values(), ignore_index=True)
df_top_5_all_cities
```

```python
produto_selecionado = df_top_5_all_cities['Produto'].iloc[0]

# Substitua 'prod-marca' pelo nome da coluna correta que identifica os produtos no seu dataframe
df_single_product = df_p[df_p['prod-descr'] == produto_selecionado]  # Ajuste o nome da coluna conforme necessário

# Calcular a correlação de Pearson para o produto selecionado
correlation_matrix_single_product = df_single_product[['prod-preco', 'spo-temp', 'spo-umid', 'spo-uv', 
                                                      'bra-temp', 'bra-umid', 'bra-uv',
                                                      'pve-temp', 'pve-umid', 'pve-uv',
                                                      'flo-temp', 'flo-umid', 'flo-uv',
                                                      'rj-temp', 'rj-umid', 'rj-uv']].corr()

# Plotar o mapa de calor para o produto selecionado
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix_single_product, annot=True, cmap='Blues', center=0)
plt.title(f'Mapa de Calor das Correlações de Pearson para {produto_selecionado}')
plt.show()
```

```python
file_path = 'D:\\Pós DS\\prep\\OPT_DATA_DETALHE.csv'

# Ler o arquivo CSV
df_d = pd.read_csv(file_path, delimiter=';', encoding='utf-8-sig')
df_d.head()
```

```python
df_filtered = df_d[df_d['prod-fps'] > 0]

# Contar o número de produtos por FPS
fps_counts = df_filtered['prod-fps'].value_counts().sort_index()

# Plotar o gráfico de barras para o número de produtos por FPS
plt.figure(figsize=(10, 6))
plt.bar(fps_counts.index, fps_counts.values, color='skyblue')
plt.xlabel('FPS')
plt.ylabel('Número de Produtos')
plt.title('Número de Produtos por FPS')
plt.xticks(fps_counts.index)
plt.show()
```

```python
binary_columns = ['toque secco', 'todos os tipos de pele', 'resistente à água', 'rápida absorção',
                  'textura leve', 'ação antioxidante', 'controle oleosidade', 'possui textura', 
                  'radicais livres', 'ácido hialurônico', 'efeito matte', 'oil free', 
                  'oil control', 'pele oleosa', 'linhas expressão', 'previne envelhecimento', 
                  'alta cobertura', 'previne rugas', 'uso diário', 'dermatologicamente testado',
                  'hydro boost', 'textura ultraleve', 'boost water', 'water gel']  

binary_columns_present = [col for col in binary_columns if col in df_d.columns]

binary_counts = df_d[binary_columns_present].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
binary_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.xlabel('Características')
plt.ylabel('Número de Produtos')
plt.title('Frequência de Produtos por Característica')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
binary_columns_present = [col for col in binary_columns if col in df_d.columns]

# Criar um único dataframe contendo as características binárias e o FPS
df_bivariate = df_d[binary_columns_present + ['prod-fps']]
df_bivariate = df_bivariate.dropna(subset=['prod-fps'])

df_bivariate = df_bivariate.melt(id_vars='prod-fps', var_name='Característica', value_name='Possui')
df_bivariate['Possui'] = df_bivariate['Possui'].replace({1: 'Sim', 0: 'Não'})

plt.figure(figsize=(15, 8))
sns.boxplot(x='Característica', y='prod-fps', hue='Possui', data=df_bivariate, palette='Set2')
plt.xticks(rotation=90)
plt.xlabel('Características Binárias')
plt.ylabel('FPS')
plt.title('Relação entre FPS e Características Binárias')
plt.legend(title='Possui a Característica', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
df_d['prod-marca'] = df_d['prod-marca'].str.strip()
df_d['prod-marca'] = df_d['prod-marca'].str.replace('(?i)isdin', 'ISDIN', regex=True)

top_brands = df_d['prod-marca'].value_counts().nlargest(10).index  

# Filtrar o dataframe para incluir apenas as top 10 marcas
df_top_brands = df_d[df_d['prod-marca'].isin(top_brands)]  
df_brand_features_top = df_top_brands.groupby('prod-marca')[binary_columns_present].sum()  

df_brand_features_top_normalized = df_brand_features_top.div(df_brand_features_top.sum(axis=1), axis=0)

plt.figure(figsize=(15, 10))
sns.heatmap(df_brand_features_top_normalized, annot=True, cmap='Blues', linewidths=0.5, linecolor='gray')
plt.title('Relação entre as Top 10 Marcas e Características Binárias')
plt.xlabel('Características Binárias')
plt.ylabel('Marca')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df_filtered = df_d[~df_d['prod-descr'].str.contains('kit', case=False, na=False)]
df_filtered.loc[:, 'total_caracteristicas'] = df_filtered[binary_columns_present].sum(axis=1)

df_top10_produtos_filtered = df_filtered.sort_values(by='total_caracteristicas', ascending=False).head(10)

df_top10_produtos_filtered = df_top10_produtos_filtered[['prod-descr', 'total_caracteristicas'] + binary_columns_present]

df_top10_produtos_filtered.reset_index(drop=True, inplace=True)
df_top10_produtos_filtered.head(20)
```

```python
plt.figure(figsize=(12, 6))
plt.barh(df_top10_produtos_filtered['prod-descr'], df_top10_produtos_filtered['total_caracteristicas'], color='skyblue', edgecolor='black')
plt.xlabel('Total de Características')
plt.ylabel('Nome do Produto')
plt.title('Top 10 Produtos com Mais Características')
plt.gca().invert_yaxis()  # Inverter o eixo y para exibir o produto com mais características no topo
plt.tight_layout()
plt.show()
```
