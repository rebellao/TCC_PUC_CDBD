#### parte 1 - unir dados em um unico arquivo CSV

```python
import os
import pandas as pd

# Unir arquivos da EPOCA
directory_path = r'D:\\Pós DS\\Dados'

# Lista para armazenar os dataframes dos arquivos que começam com "EPOCA"
epoca_dataframes = []

# Percorre todos os arquivos no diretório
for filename in os.listdir(directory_path):
    if filename.startswith("EPOCA") and filename.endswith(".csv"):
        # Cria o caminho completo para o arquivo
        file_path = os.path.join(directory_path, filename)
        
        try:
            # Lê o arquivo CSV usando o delimitador ";"
            df = pd.read_csv(file_path, delimiter=';', encoding='mbcs')
            
            epoca_dataframes.append(df)
        
        except Exception as e:
            print(f"Erro ao ler o arquivo {file_path}: {e}")

combined_epoca_df = pd.concat(epoca_dataframes, ignore_index=True)

combined_epoca_df.to_csv(r'D:\\Pós DS\\prep\\full_epoca_data.csv', index=False, sep=';', encoding='utf-8-sig')
```

```python
import re

# Função para extrair o fator de proteção FPS do campo 'prod-descr'
def extrair_fps(descricao):
    match = re.search(r'FPS\s*(\d+)', descricao, re.IGNORECASE)
    if match:
        return int(match.group(1))
    return None

# Criar a nova coluna 'prod_fps' aplicando a função na coluna 'prod-descr'
df_det_ordenado['prod_fps'] = df_det_ordenado['prod_descricao'].apply(extrair_fps)
```

```python
df_det_ordenado['prod_fps'] = df_det_ordenado['prod_fps'].astype('Int64')
df_det_ordenado.head()
```

```python
df_det_ordenado['prod_fps'] = df_det_ordenado['prod_fps'].fillna(0).astype(int)
df_det_ordenado.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import pandas as pd
import nltk

nltk.download('stopwords')
portuguese_stopwords = stopwords.words('portuguese')

corpus = df_det_unique['prod_detalhe'].fillna("").tolist()  # Preencher valores ausentes com string vazia

vectorizer = TfidfVectorizer(ngram_range=(2, 3), stop_words=portuguese_stopwords, max_features=50)
X = vectorizer.fit_transform(corpus)

# Obter as palavras-chave e expressões mais importantes
feature_names = vectorizer.get_feature_names_out()
importance = X.toarray().sum(axis=0)

# Criar um dataframe com as palavras-chave e sua importância
keywords_df = pd.DataFrame({'expressao': feature_names, 'importancia': importance})
keywords_df = keywords_df.sort_values(by='importancia', ascending=False)

print(keywords_df)
```

```python
df_comentarios.loc[:, 'prod_avaliacao'] = df_comentarios['prod_avaliacao'].astype(str).str.replace(',', '.').astype(float)
df_comentarios.head()
```

```python
df_comentarios.loc[:, 'prod_comentario'] = df_comentarios['prod_comentario'].str.replace('\n', ' ', regex=True)
df_comentarios.head()
```

```python
df_comentarios = df_comentarios.drop_duplicates()
df_comentarios.head()
```

```python
df_comentarios['prod_comentario'] = df_comentarios['prod_comentario'].fillna("").astype(str)

# Função para quebrar os comentários
def quebrar_comentarios(row):
    comentarios = row['prod_comentario'].split('Esta avaliação foi útil?')
    return [(row['prod_descricao'], row['prod_marca'], row['prod_avaliacao'], comentario.strip()) for comentario in comentarios if comentario.strip()]

comentarios_expandidos = df_comentarios.apply(quebrar_comentarios, axis=1).explode().dropna()
df_comentarios_quebrados = pd.DataFrame(comentarios_expandidos.tolist(), columns=['prod_descricao', 'prod_marca', 'prod_avaliacao', 'prod_comentario'])

df_comentarios_quebrados.head()
```

```python
df_coment_ord = df_comentarios_quebrados.sort_values(by="prod_descricao", ascending=True)
df_coment_ord.head()
```

```python
def remover_stopwords(comentario):
    palavras = comentario.split()
    palavras_sem_stopwords = [palavra for palavra in palavras if palavra.lower() not in portuguese_stopwords]
    return ' '.join(palavras_sem_stopwords)

# Aplicar a função de remoção de stopwords na coluna 'prod-comentario'
df_coment_ord['prod_comentario'] = df_coment_ord['prod_comentario'].apply(remover_stopwords)
df_coment_ord.head()
```

```python
df_coment_ord = df_coment_ord.drop_duplicates()
df_coment_ord.head(20)
```

```python
import re

def remover_datas(comentario):
    return re.sub(r'\b\d{1,2}\s+[a-zA-Zçáéíóúãõ]+(?:\s+de\s+)?\d{4}\b|\b\d{2}/\d{2}/\d{4}\b', '', comentario, flags=re.IGNORECASE)

df_coment_ord['prod_comentario'] = df_coment_ord['prod_comentario'].apply(remover_datas)
df_coment_ord.head()
```
