```python
file_path = 'D:\\Pós DS\\prep\\FULL_DATA_ATR.csv'

# Ler o arquivo CSV
df_atr = pd.read_csv(file_path, delimiter=';', encoding='utf-8-sig')
df_atr.head()
```

```python
df_atr = df_atr.drop(columns=['importancia'])
df_atr = df_atr.drop_duplicates()
```

```python
df_atr.head(20)
```

```python
# para cada linha de atributos, criar coluna no df de produtos e detalhes, se possuir o atributo, marcar com 1, senao 0
for expressao in df_atr['expressao']:
    df_det[expressao] = df_det['prod-detalhe'].apply(lambda x: 1 if expressao in x else 0)
```
