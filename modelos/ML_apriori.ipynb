```python
import pandas as pd

file_path = 'D:\\Pós DS\\prep\\OPT_DATA_DETALHE.csv'

# Ler o arquivo CSV
df = pd.read_csv(file_path, delimiter=';', encoding='utf-8-sig')
df.head()
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from IPython.display import display

# Suponha que o dataset seja chamado de 'df'
df_bin = df.drop(columns=['prod-descr', 'prod-marca', 'prod-detalhe', 'prod-fps'])

# Convertendo as colunas para tipo bool (True/False)
df_bin = df_bin.astype(bool)

# Aplicando o algoritmo Apriori com suporte mínimo
frequent_itemsets = apriori(df_bin, min_support=0.05, use_colnames=True)

# Gerando regras de associação com confiança mínima ajustada
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Configurando pandas para exibir todas as colunas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Configurando para exibir com 3 casas decimais
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Exibindo as regras no formato de tabela
display(rules)
```
